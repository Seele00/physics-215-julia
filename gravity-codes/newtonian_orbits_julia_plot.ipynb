```julia
include("newtonian_orbits_leapfrog_julia.jl")
```

```julia
#initializing step
r = [1 0]
v = [0 0.5]
a = gravforce(r)
energy = totalenergy(r, v)

dt = 0.01
T = 7
```

```julia
t = 0
x = append!(Vector{Float64}(), r[1])
y = append!(Vector{Float64}(), r[2])

@timev while t < T
    r, v, a = leapfrog(r, v, a, dt)
    t += dt
    
    append!(x, r[1])
    append!(y, r[2])
end
```

```julia
using Plots
gr()

@timev scatter(x, y, title="Leapfrog-integrated Newtonian orbit", label=false, aspect_ratio=:equal)
```
